## Histograms

Histograms are most commonly used to show frequency distributions (how many items are there for each category)

This is useful to see the shape of a data distribution.

This is the code to create a histogram on genres.

```python
sns.histplot(df, x='genre')
plt.xticks(rotation = 45)
```

Did you see a lot of unwanted texts before the plot itself? We'll discuss on a way to tackle that.

Notice the blandness and the labels. Submitting plots like these is not the way to go. Submitting your plots in the default formatting style is like creating a presentation without a theme (in basic white).People can get lose interest immediately. In a business setting, people might be tired of seeing plots in the default format.

You also need to format plots because:
    * Your company might have a style that is associated with them and they expect it to be followed
    * You can highlight trends so as to frame your argument. (Think of an example in the chart above where you work for a movie studio that only does Horror Movies, you can highlight the horror part of the histogram in a different color to draw interest)



You need the following:

* Proper Labels
* Chart Titles
* A different color scheme 

```python
sns.set(rc={"figure.figsize":(16,9)})#Width = 16 Height = 9
#palette = {"Adventure":"tab:purple"}
_ = sns.histplot(df, x='genre',hue='genre')
_ = plt.xticks(rotation = 45)
_ = plt.xlabel("Genre of Movies")
_ = plt.ylabel("Number of Movies in Dataset")
_ = plt.title("Distribution of Movie Genres in the Dataset")
```

## Frame of Reference

Sometimes plotting things as is does not give you the whole picture. You might have to work on it a bit to get the message across. This does not mean manipulating the chart to force a trend, but certain aspects can throw the entire chart out of whack.

```python
_ = sns.scatterplot(data = df, x = 'IMDB_score', y =  'budget', hue= "genre")
_ = plt.xlabel("IMDB Score")
_ = plt.ylabel("Budget")
_ = plt.title("Budget for Movies across IMDB Ratings")
```

Notice that the chart above has all the points scattered close to the x-axis. This might give the impression that the budget for all these movies are similiar. However, if you notice, it is one point (outlier) that is throwing the entire axis out of whack. Normalising it by applying a logarithmetic scale should give a better picture

```python
_ = sns.scatterplot(data = df, x = 'IMDB_score', y =  'budget', hue = 'genre').set_yscale("log")
_ = plt.xlabel("IMDB Score")
_ = plt.ylabel("Budget")
_ = plt.title("Budget for Movies across IMDB Ratings")
```

Some would say that the colors used in this chart is borderline unreadable, but I guess I'm just at the edge of what is readable (number of categories is too high). To give you an idea of what is too high:

```python
_ = sns.scatterplot(data = df, x = 'IMDB_score', y =  'budget', hue = 'country').set_yscale("log")
_ = plt.xlabel("IMDB Score")
_ = plt.ylabel("Budget")
_ = plt.title("Budget for Movies across IMDB Ratings")
```

Which is France? Which is USA? Which is Kenya?

```python
_ = df.groupby('title_year').budget.min().plot(label = 'min', c = 'green').set_yscale("log")
_ = df.groupby('title_year').budget.max().plot(label = 'max', c = 'orange').set_yscale("log")
_ = plt.legend(loc='best')
_ = plt.xlabel("Year of Release")
_ = plt.ylabel("Budget")
_ = plt.title("Budget for Movies across Years (Min & Max)")
```

```python
_ = df.groupby('title_year').gross.min().plot(label = 'min', c = 'green').set_yscale("log")
_ = df.groupby('title_year').gross.max().plot(label = 'max', c = 'orange').set_yscale("log")
_ = plt.legend(loc='best')
_ = plt.xlabel("Year of Release")
_ = plt.ylabel("Budget")
_ = plt.title("Budget for Movies across Years (Min & Max)")
```

# Manipulating the Dataset for Visualisations

The earlier plots showed the min and max budgets/gross earnings for movies across the years. What if we wanted to introduce a calculated field, say profit?

For that we need to create a column in the original dataset

```python
df["nett"] = df["gross"] - df["budget"] # Creating the column
```

```python
df.head() #Checking to see if the column came out well
```

We can create a groupby object instead of typing it out each time.

```python
gby = df.groupby("title_year") 
```

```python
gby.nett.mean().plot(c = "purple", lw = 4) #lw = linewidth, c = color
_ = plt.xlabel("Year of Release")
_ = plt.ylabel("Nett Profit/Loss")
_ = plt.title(" Average Nett Profit/Loss for Movies across the Years")
```
